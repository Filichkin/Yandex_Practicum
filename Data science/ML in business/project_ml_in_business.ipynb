## Создание и обучение моделей

```python
def region_predict(data):
    model_lr = LinearRegression()
    X = data.drop(['product', 'id'], axis = 1)
    y = data['product']
    X_train, X_valid, y_train, y_valid = train_test_split(X, y,
                                                          random_state = RANDOM_STATE,
                                                          test_size = TEST_SIZE
                                                         )
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_valid_scaled = scaler.transform(X_valid)
    model_lr.fit(X_train_scaled, y_train)
    predictions = model_lr.predict(X_valid_scaled)
    rmse_score = mean_squared_error(y_valid, predictions, squared = False)
    mean_predicted_product = predictions.mean() 
    return rmse_score, mean_predicted_product, predictions, y_valid
```

```python
rmse_0, mean_pred_prod_0, pred_0, y_valid_0 = region_predict(geo_data_0)
rmse_1, mean_pred_prod_1, pred_1, y_valid_1 = region_predict(geo_data_1)
rmse_2, mean_pred_prod_2, pred_2, y_valid_2 = region_predict(geo_data_2)
print('Регион 0:\n\n',
    f'RMSE = {rmse_0:.2f}, средний запас предсказанного сырья = {mean_pred_prod_0:.2f}', '\n')
print('Регион 1:\n\n',
    f'RMSE = {rmse_1:.2f}, средний запас предсказанного сырья = {mean_pred_prod_1:.2f}', '\n')
print('Регион 2:\n\n',
    f'RMSE = {rmse_2:.2f}, средний запас предсказанного сырья = {mean_pred_prod_2:.2f}', '\n')
```

### Суммирование целевого значения объёма сырья и расчет прибыли

```python
def target_sum(target, predictions):
    predicted_sorted = predictions.sort_values(ascending=False)
    selected = target[predicted_sorted.index][:BEST_POINTS]
    return selected.sum() * PRICE - BUDGET
```

```python
def bootstrap_geo(target, probabilities):
    state = np.random.RandomState(12345)
    values=[]
    for i in range(1000):
        target_subsample = target.sample(replace=True,\
            random_state=state, n=MAX_POINTS)

        probs_subsample = probabilities[target_subsample.index]
        values.append(target_sum(target_subsample, probs_subsample))

    values = pd.Series(values)
    lower = values.quantile(0.025)
    upper = values.quantile(0.975)
    mean = values.mean()
    risk = st.percentileofscore(values, 0)

     # Отображение таблицы
    table = PrettyTable()
    table.field_names = ['Показатель', 'Значение']
    table.add_row(['Средняя прибыль', f'{mean/1e6:.1f} млн'])
    table.add_row(['Риск', f'{risk:.2}%'])
    table.add_row(['Доверительный интервал', f'{lower/1e6:.1f} : {upper/1e6:.1f} млн'])
    print(table)

    sns.histplot(values, color='#9BCD9B', alpha=0.5,  element='step')
    plt.xlabel('Прибыль')
    plt.axvline(x=mean, color='black', ls='-', lw=2, label='Средняя прибыль')
    plt.axvline(x=lower, color='r', ls=':', lw=3, label='95%-й доверительный интервал')
    plt.axvline(x=upper, color='r', ls=':', lw=3)
    plt.legend(bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=2)
    plt.show()

bootstrap_geo(y_valid_0, pd.Series(pred_0, index=y_valid_0.index))
```
